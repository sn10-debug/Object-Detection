```python
import cv2
import mediapipe as mp
import numpy as np
mp_drawing=mp.solutions.drawing_utils
mp_pose=mp.solutions.pose
```

```python
def calculate_angle(a,b,c):
    a=np.array(a)
    b=np.array(b)
    c=np.array(c)
    
    radians=np.arctan2(c[1]-b[1],c[0]-b[0])-np.arctan2(a[1]-b[1],a[0]-b[0])
    angle=np.abs((radians*180)/np.pi)
    if angle>180:
        angle=360-angle
    return angle
```

```python
capture=cv2.VideoCapture(0)
landmarks=[]
coords=[]
with mp_pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5) as pose:
    left_stage="down"
    left_counter=0
    right_stage="down"
    right_counter=0
    
    while capture.isOpened():
        ret,frame=capture.read()
        
        
#         Recolour image to RGB
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image.flags.writeable=False
        
#         Make detection
        results=pose.process(image)
    
        
        image.flags.writeable=True
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
    
        
        

        try:
            landmarks=results.pose_landmarks.landmark
    #             coords=[landmarks[12],landmarks[14],landmarks[11],landmarks[13]]
    #             if abs(coords[0].y-coords[1].y)<=0.01 or abs(coords[0].x-coords[1].x)<=0.01:
    #                 print("Right Arm is Straight")

    #             if abs(coords[2].y-coords[3].y)<=0.01 or abs(coords[2].x-coords[3].x)<=0.01:
    #                 print("Left Arm is Straight")
            left_shoulder=[landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            left_elbow=[landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            left_wrist=[landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
            left_hip=[landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            right_shoulder=[landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            right_elbow=[landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            right_wrist=[landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
            right_hip=[landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
            
            angle1=calculate_angle(left_shoulder,left_elbow,left_wrist)
            angle2=calculate_angle(left_hip,left_shoulder,left_elbow)
            
            angle3=calculate_angle(right_shoulder,right_elbow,right_wrist)
            angle4=calculate_angle(right_hip,right_shoulder,right_elbow)
            cv2.putText(image,str(angle1),tuple(np.multiply(left_elbow,[640,480]).astype(int)),

                       cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
            cv2.putText(image,str(angle2),tuple(np.multiply(left_shoulder,[640,480]).astype(int)),

                       cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
            cv2.putText(image,str(angle3),tuple(np.multiply(right_elbow,[640,480]).astype(int)),

                       cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
            cv2.putText(image,str(angle4),tuple(np.multiply(right_shoulder,[640,480]).astype(int)),

                       cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
            
            if angle1>160:
                left_stage="down"
            
            if angle1<30 and right_stage=="down":
                left_stage="up"
                left_counter=right_counter+1
                print("Left Curls : "+str(left_counter))
            
    
            if angle3>160:
                right_stage="down"
            
            if angle3<30 and left_stage=="down":
                right_stage="up"
                right_counter=left_counter+1
                print("Right Curls : "+str(right_counter))
            
                

        except:
            pass
#         Recolour image to BGR
        
#         Render detection
        mp_drawing.draw_landmarks(image,results.pose_landmarks,mp_pose.POSE_CONNECTIONS,mp_drawing.DrawingSpec(color=(245,117,66),thickness=2,circle_radius=2),
                                 mp_drawing.DrawingSpec(color=(19, 41, 240),thickness=2,circle_radius=2))
        
#         The colour given here is in BGR format in the tuple
        
#         This POSE_connection represents the connections between the links

        cv2.imshow('frame',image)
        if cv2.waitKey(10) & 0xFF==ord('q'):
            break

    capture.release()
    cv2.destroyAllWindows()
```

```python
def build_model(hp):
    model=keras.Sequential([
        keras.layers.Conv2D(
        filters=hp.Int('conv_1_filter',min_value=32,max_value=128,step=16),
#             This will select a range of number of filters that is minimum 
#             number and max no. is 32 and 128 resp.
            
            kernel_size=hp.Choice('conv_1_kernel',values=[3,5]),
            activation='relu',
            input_shape=(28,28,1)
            
        ),
        keras.layers.Conv2D(
        filters=hp.Int('conv_2_filter',min_value=32,max_value=64,step=16),
            kernel_size=hp.Choice('conv_2_kernel',values=[3,5]),
            activation='relu'
        ),
        keras.layers.Flatten(),
        keras.layers.Dense(
        units=hp.Int('dense_1_units',min_value=32,max_value=128,step=16),
            activation='relu'
        ),
        keras.layers.Dense(10,activation='softmax')
    ])
    
    model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning_rate',values=[1e-2,1e-3])),
                  loss="sparse_categorical_crossentropy",
                  metrics=['accuracy'])
    
    return model
```
